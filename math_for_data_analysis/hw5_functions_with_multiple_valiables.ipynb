Посчитайте частные производные функций:

```python
import sympy as sym
import numpy as np
from math import sin,cos
```

```python
x=sym.Symbol('x')
y=sym.Symbol('y')
f1 = 2*x**2*y**3+1/x+y**2*x+7
f2 = x**2*y-sin(x*y)+cos(x**2)+6*y
```

```python
print("1 фунция")
print(f"Частная производная по x: {sym.diff(f1,x)}")
print(f"Частная производная по y: {sym.diff(f1,y)}")
print("2 фунция")
print(f"Частная производная по x: {sym.diff(f2,x)}")
print(f"Частная производная по y: {sym.diff(f2,y)}")
```

###Задание 2
Вычислите градиент функции f1 в точке (1,2)

```python
def grad_1(x, y):
    f1 = 2*x**2*y**3+1/x+y**2*x+7
    dx = sym.diff(f1,x)
    dy = sym.diff(f1,y)
    return [dx, dy]
grad_1(x,y)
```

```python
np.gradient
```

```python
M = {x:1, y:2}
grad_M = []

for f in grad_1(x,y):
  grad_M.append(f.subs(M).evalf()) 
print(f"Значение градиента в точке равно: {grad_M}")
```

```python
# А теперь попробуем тот же график построить с помощью другой библиотеки
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

#Make data
x1 = np.arange(-10, 10, 0.01)
x2 = np.arange(-10, 10, 0.01)
x1, x2 = np.meshgrid(x1, x2)
Z = x2**2 - ((x1**2)*x2) + (2*x1*x2)

surf = ax.plot_surface(x1, x2, Z, linewidth=1)
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

fig.set_figwidth(12)    
fig.set_figheight(10) 
plt.show()
```
