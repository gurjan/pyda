Посчитайте значение первой и второй производной функции: x**5 + 4*sym.sin(2*x) + sym.cos(3*x+3)

В точке x = 1.

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
# simvolnaja peremennaja
x = sym.Symbol('x')
```

```python
f_x=x**5 + 4*sym.sin(2*x) + sym.cos(3*x+3)
f_x
```

```python
f_first = f_x.diff(x)
f_first
```

```python
f_second = f_first.diff(x)
f_second
```

```python
print(f"значение первой производной функции {f_first.subs(x,1)} : ", f_first.evalf(subs={x: 1}))
print(f"значение второй производной функции {f_second.subs(x,1)} : ", f_second.evalf(subs={x: 1}))
```
